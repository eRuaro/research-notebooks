```python
import arrow
import requests
import json

stormglass_api = '51336722-4f14-11ec-b247-0242ac130002-513367c2-4f14-11ec-b247-0242ac130002'
weatherapi_key = '995333b3c1334b7da0104356212711'

date = '2021-12-10'
start = arrow.get(date)
q = 'Makati'
lat = 14.57
lng = 121.03
params = [
    'humidity',
    'precipitation', 
    'pressure',
]

weather_response = requests.get(
    'https://api.weatherapi.com/v1/history.json',
    params={
        'key': weatherapi_key,
        'q': q,
        'dt': date,
    },
)

response = requests.get(
    'https://api.stormglass.io/v2/weather/point',
    params={
        'lat': lat,
        'lng': lng,
        'params': ','.join(params), 
        'start': start.to('Asia/Manila').timestamp(), # convert to UTC
        'end': start.shift(hours=23).to('Asia/Manila').timestamp(),
    },
    headers={
        'Authorization': stormglass_api
    }
)

history_weather_data = weather_response.json()['forecast']['forecastday'][0]['hour']
stormglass_data = response.json()['hours']

for i in stormglass_data:
    humidity = i['humidity']['noaa']
    precipitation = i ['precipitation']['noaa']
    pressure = i['pressure']['noaa']
    list_relative_humidity.append(humidity)
    list_precipitation.append(precipitation)
    list_air_pressure.append(pressure)

for i in history_weather_data:
    name = q
    region = q + " City"
    country = "Philippines"
    longitude = lng
    latitude = lat
    time = i['time']
    temp_c = i['temp_c']
    is_day = i['is_day']
    condition = i['condition']['text']
    wind_kph = i['wind_kph']
    wind_degree = i['wind_degree']
    wind_dir = i['wind_dir']
    cloud = i['cloud']
    feelslike_c = i['feelslike_c']
    will_it_rain = i['will_it_rain']
    chance_of_rain = i['chance_of_rain']
    list_name.append(name)
    list_region.append(region)
    list_country.append(country)
    list_longitude.append(longitude)
    list_latitude.append(latitude)
    list_time.append(time)
    list_temp_c.append(temp_c)
    list_is_day.append(is_day)
    list_condition.append(condition)
    list_wind_kph.append(wind_kph)
    list_wind_degree.append(wind_degree)
    list_wind_dir.append(wind_dir)
    list_cloud.append(cloud)
    list_feelslike_c.append(feelslike_c)
    list_chance_of_rain.append(chance_of_rain)
    list_will_it_rain.append(will_it_rain)
```
