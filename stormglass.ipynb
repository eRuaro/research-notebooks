```python
import arrow
import requests
import json

stormglass_api = '51336722-4f14-11ec-b247-0242ac130002-513367c2-4f14-11ec-b247-0242ac130002'

start = arrow.get('2021-12-08')

params = [
    'humidity',
    'precipitation', 
    'pressure',
]
response = requests.get(
    'https://api.stormglass.io/v2/weather/point',
    params={
        'lat': 14.57,
        'lng': 121.03,
        'params': ','.join(params), 
        'start': start.to('Asia/Manila').timestamp(), # convert to UTC
        'end': start.shift(hours=23).to('Asia/Manila').timestamp(),
    },
    headers={
        'Authorization': stormglass_api
    }
)

stormglass_data = json.dumps(response.json(), indent=4)
print(stormglass_data)
```
