## Stormglass

```python
import arrow
import requests
import json

stormglass_api = '51336722-4f14-11ec-b247-0242ac130002-513367c2-4f14-11ec-b247-0242ac130002'

start = arrow.get('2021-12-10')

params = [
    'airTemperature',
    'humidity',
    'precipitation', 
    'pressure',
    'windDirection', 
    'windSpeed',
]
response = requests.get(
    'https://api.stormglass.io/v2/weather/point',
    params={
        'lat': 14.58,
        'lng': 121.1169444,
        'params': ','.join(params), 
        'start': start.to('UTC').timestamp(), # convert to UTC
        'end': start.shift(days=1).to('UTC').timestamp(),
    },
    headers={
        'Authorization': stormglass_api
    }
)

stormglass_data = json.dumps(response.json(), indent=4)
print(stormglass_data)
```
