```python
list_name = []
list_region = []
list_country = []     
list_longitude = []
list_latitude = []
list_time = []# YYYY-MM-DD HH:MM 
list_temp_c = []
list_is_day = []
list_condition = []
list_air_pressure = [] # hPa
list_relative_humidity = [] # relative humidity in percent
list_precipitation = [] # precipitation in mm
list_wind_kph = [] # wind speed in km/h
list_wind_degree = [] # 0-360
list_wind_dir = [] # N, NE, E, SE, S, SW, W, NW
list_cloud = [] # cloudiness in percent
list_feelslike_c = [] # feels like temperature in degrees C
list_will_it_rain = []
list_chance_of_rain = []
```

```python
import arrow
import requests
import json
import pandas as pd

stormglass_api = '51336722-4f14-11ec-b247-0242ac130002-513367c2-4f14-11ec-b247-0242ac130002'
weatherapi_key = '995333b3c1334b7da0104356212711'

date = '2022-02-19'
start = arrow.get(date)
q = '10.7, 122.56'
lat = 10.7
lng = 122.56
params = [
    'humidity',
    'precipitation', 
    'pressure',
]

weather_response = requests.get(
    'https://api.weatherapi.com/v1/history.json',
    params={
        'key': weatherapi_key,
        'q': q,
        'dt': date,
    },
)

response = requests.get(
    'https://api.stormglass.io/v2/weather/point',
    params={
        'lat': lat,
        'lng': lng,
        'params': ','.join(params), 
        'start': start.to('Asia/Manila').timestamp(), # convert to UTC
        'end': start.shift(hours=23).to('Asia/Manila').timestamp(),
    },
    headers={
        'Authorization': stormglass_api
    }
)

history_weather_data = weather_response.json()['forecast']['forecastday'][0]['hour']
stormglass_data = response.json()['hours']

for i in stormglass_data:
    humidity = i['humidity']['noaa']
    precipitation = i ['precipitation']['noaa']
    pressure = i['pressure']['noaa']
    list_relative_humidity.append(humidity)
    list_precipitation.append(precipitation)
    list_air_pressure.append(pressure)

for i in history_weather_data:
    name = "Iloilo"
    region = name + " City"
    country = "Philippines"
    longitude = lng
    latitude = lat
    time = i['time']
    temp_c = i['temp_c']
    is_day = i['is_day']
    condition = i['condition']['text']
    wind_kph = i['wind_kph']
    wind_degree = i['wind_degree']
    wind_dir = i['wind_dir']
    cloud = i['cloud']
    feelslike_c = i['feelslike_c']
    will_it_rain = i['will_it_rain']
    chance_of_rain = i['chance_of_rain']
    list_name.append(name)
    list_region.append(region)
    list_country.append(country)
    list_longitude.append(longitude)
    list_latitude.append(latitude)
    list_time.append(time)
    list_temp_c.append(temp_c)
    list_is_day.append(is_day)
    list_condition.append(condition)
    list_wind_kph.append(wind_kph)
    list_wind_degree.append(wind_degree)
    list_wind_dir.append(wind_dir)
    list_cloud.append(cloud)
    list_feelslike_c.append(feelslike_c)
    list_chance_of_rain.append(chance_of_rain)
    list_will_it_rain.append(will_it_rain)
```

```python
data = {
    'name': list_name,
    'region': list_region,
    'country': list_country,
    'longitude': list_longitude,
    'latitude': list_latitude,
    'time': list_time,
    'temp_c': list_temp_c,
    'is_day': list_is_day,
    'condition': list_condition,
    'wind_kph': list_wind_kph,
    'wind_degree': list_wind_degree,
    'wind_dir': list_wind_dir,
    'air_pressure': list_air_pressure,
    'precipitation': list_precipitation,
    'relative_humidity': list_relative_humidity,
    'cloud': list_cloud,
    'feelslike_c': list_feelslike_c,
    'will_it_rain': list_will_it_rain,
    'chance_of_rain': list_chance_of_rain,
}
df = pd.DataFrame(data=data)

df.tail()
```
